```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('companies.csv')
```

```python
group_by_year = df.loc[:, ['year',  'profit']].groupby('year')

#Select the averages for each year
avgs = group_by_year.mean()

# X will be the years 
x = avgs.index 

# Y will be the increase in mean profit for the year
y1 = avgs.profit
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x, y)
    ax.margins(x=0, y=0)
```

```python
fig, ax = plt.subplots()
plot(x, y1, ax, 'Increase in Mean Revenue', 'Profit in Millions')
```
